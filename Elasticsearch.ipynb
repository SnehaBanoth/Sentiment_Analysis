```python
#Create index
from elasticsearch import Elasticsearch

es = Elasticsearch([{'host': '34.90.37.18', 'port': 9200}])
es.indices.create(index="sentiment_prediction_7", ignore=400)
```

```python
from kafka import KafkaConsumer 
#load data from kafka into elasticsearch
import json
import uuid ## to generate randum id number if you dont have an id code in your data

while True: 
    consumer = KafkaConsumer(bootstrap_servers='34.90.212.110:9092',
                                     auto_offset_reset='latest',
                                     consumer_timeout_ms=1000)
    consumer.subscribe(['predictions_7'])

    i = 0
    for msg in consumer:

        print(i)
        es.index(index="sentiment_prediction_7", 
                 doc_type = "label_prediction", 
                 id = uuid.uuid1(), 
                 body = json.loads(msg.value.decode('utf-8')))
        i = i + 1
```
